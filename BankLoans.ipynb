```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df_loans= pd.read_csv('C:/Users/Massa/Desktop/DataScience/loans.csv')
print(df_loans[['B Salary','B commitments ','B AGE','S salary','S commitments ','S AGE']])
def count_loans(data):
    counter_c = 0
    counter_p = 0
    counter_h = 0
    for i in range (0 , 9673):
        if (((data['S salary'][i]/2) - data['S commitments '][i] > 0) and (data['S AGE'][i] - 60 > 0)) :
            if ((data['B Salary'][i]/2) - data['B commitments '][i] > 0) :
                if (data['B AGE'][i] - 60 > 0) :
                    loan_value = (60 - data['B AGE'][i]) * (12)* ((data['B Salary'][i]/2) - data['B commitments '][i])
                    multiple = abs(loan_value / data['B Salary'][i])
                    if multiple >= 50 :
                        counter_c = counter_c + 1
                        counter_p = counter_p + 1
                        counter_h = counter_h + 1
                    else:
                        if multiple >= 30 :
                            counter_c = counter_c + 1
                            counter_p = counter_p + 1
                        else:
                                    if(multiple >= 10) :
                                        counter_p = counter_p + 1
    return counter_c , counter_p , counter_h ;

car , personal , house = count_loans(df_loans)
print ("===========================================================================================================")
print("the number of car loans is :", car)
print("the number of house loans is :", house)
print("the number of personal loans is :",personal)
```

```python
def abc(data):
    """Compute ECDF for a one-dimensional array of measurements."""

    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.mean(data)
    y = np.median(data)
    z = np.std(data)

   

    return x, y , z
```

```python
x_OH, y_OH, Z_OH = abc(df_loans['B Salary'])
x_PA, y_PA, Z_PA = abc(df_loans['B AGE'])
x_FL, y_FL, Z_FL= abc(df_loans['B commitments '])

print(x_OH , y_OH ,Z_OH)
print(x_PA , y_PA ,Z_PA)
print(x_FL , y_FL ,Z_FL)
```
